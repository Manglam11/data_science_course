### Step 3 — Fix Column Names

```python
df.columns.to_list()
```

```python
df.columns = df.columns.str.replace("\xa0", " ", regex=False).str.strip()
```

```python
df.columns.to_list()
```

### Step 4 — Drop Useless Column `Ref.`

```python
df.drop(columns="Ref.", inplace=True)
print(df.columns.to_list())
print()
print(df.shape)
```

### Step 5 — Clean Money Columns 
#### `Actual gross, Average gross, Adjusted gross (in 2022 dollars)`

```python
for col in ["Actual gross", "Average gross", "Adjusted gross (in 2022 dollars)"]:
    df[col] = df[col].str.replace(r"[^0-9]", "", regex=True).astype(int)

print(df[["Actual gross", "Average gross", "Adjusted gross (in 2022 dollars)"]].dtypes)
print(df[["Actual gross", "Average gross", "Adjusted gross (in 2022 dollars)"]].head())
```

### Step 6 — Clean `Peak` & `All Time Peak` Columns

```python
df["Peak"] = df["Peak"].str.replace(r"\[.*?\]", "", regex=True).str.strip()
df["All Time Peak"] = df["All Time Peak"].str.replace(r"\[.*?\]", "", regex=True).str.strip()

print(df[["Peak", "All Time Peak"]])
```

### Step 7 — Clean `Tour title` Column

```python
df["Tour title"] = df["Tour title"].str.replace(r"\[.*?\]", "", regex=True)
df["Tour title"] = df["Tour title"].str.replace(r"[^a-zA-Z0-9\s\-&]", "", regex=True).str.strip()
df["Tour title"]
```

### Step 8 — Clean `Year(s)` Column

```python
df["Year(s)"] = df["Year(s)"].str.extract(r"(\d{4})").astype(int)
df["Year(s)"]
```
