## Python Sets
### 1. Basic

```python
s1 = {1, 2, 3, 4, 5}
print(s1)
print(type(s1))
```

### 2. Duplicate Removal

```python
nums = [1, 2, 2, 3, 4, 4, 5]
print(set(nums))
```

### 3. Add Elements

```python
s1.add(7)
print(s1)
s1.update([8, 9]) #for multiple element
print(s1)
```

### 4. Remove Elements

```python
s1.remove(3)
# s1.remove(10) will throw error
s1.discard(10) # will not
```

### 5. Loop Through Set

```python
s = {"apple", "banana", "cherry"}
for element in s:
    print(element)
```

### 6. Membership Check

```python
if "banana" in s: print(True)
```

### 7. Set Operations


```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(A | B) # Union
print(A & B) # Intersection
print(A - B) # Difference
print(A ^ B) # Symmetric Difference
```

### 8. Common Elements Between Lists

```python
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]

print(set(list1) & set(list2))
```

### 9. Unique Elements from Two Lists

```python
print(set(list1) | set(list2))
```

### 10. Subset Check

```python
A = {1, 2}
B = {1, 2, 3, 4}

print(A.issubset(B))
```

### 11. Disjoint Sets

```python
A = {1, 2}
B = {3, 4}
print(A.isdisjoint(B))
```

### 12. Count Unique Words

```python
sentence = "python is easy and python is powerful"
print(set(sentence))
print(len(set(sentence)))
```

### 13. Challenge Question

```python
python_students = {"Aman", "Ravi", "Sita"}
java_students = {"Ravi", "Sita", "John"}

print(f"Students in both courses: {python_students & java_students}")
print(f"Students only in Python: {python_students - java_students}")
print(f"Students in either courses: {python_students | java_students}")
print(f"Students only in one course: {python_students ^ java_students}")
```
