### Python Dictionaries
### 1. Basics

```python
dic_1 = {
    "name": "manglam",
    "age": 21,
    "course": "FSDS"
}
print(dic_1.keys())
print(dic_1.values())
print(dic_1.items())

print(dic_1.get("name"))
# print(dic_1.pop("name"))

dic_1["city"] = "raipur"
print(dic_1)

dic_1["age"] = 60
print(dic_1)

dic_1.pop("course")
print(dic_1)

del dic_1["age"]
print(dic_1)
```

### 2. Understanding Operations

```python
d = {"a": 10, "b": 20, "c": 30}

for key in d:
    print(key)

for key in d:
    print(d[key])

coutn = 0
for key in d:
    coutn += 1
print(coutn)

# is "b" in d

d1 = {"x": 1}
d2 = {"y": 2}

new_dic = d1 | d2
print(new_dic)

my_dic = dict([("fruit1", "banana"), ("fruit2", "apple")])
print(my_dic)
```

### 3. Logic Building

```python
s = "banana"
s_dict = {char: s.count(char) for char in s}
print(s_dict)
```

```python
marks = {"math": 90, "science": 85, "english": 95}
print(max(marks, key= marks.get))
```

```python
square_keys = {n: n**2 for n in range(1,6)}
print(square_keys)
```

```python
non_reverse_dic = {"a": 1, "b": 2}
reverse_dic = {non_reverse_dic[i]:i for i in non_reverse_dic}
print(reverse_dic)
```

```python
d = {"a": 10, "b": 25, "c": 5}
new_d = {i:d[i] for i in d if d[i] > 20}
print(new_d)

soreted_d = dict(sorted(d.items(), key=lambda x: x[1]))
print(soreted_d)
```

```python
d1 = {"a": 10, "b": 25, "c": 5}
d2 = {"a": 20, "b": 55, "d": 5}

d1_key = set(d1.keys())
d2_key = set(d2.keys())
print(d1_key & d2_key)
```

```python
student_dict = {"Ram": 85, "Shyam": 90, "Amit": 78}

topper_name = ""
topper_marks = 0
for i in student_dict:
    if student_dict[i] > topper_marks:
        topper_marks = student_dict[i]
        topper_name = i

print(topper_name)
print(topper_marks)
```

```python
s = "I love python python is easy"
words = s.split()
word_dict = {}

for word in words:
    if word in word_dict:
        word_dict[word] +=1
    else:
        word_dict[word] = 1
print(word_dict)
```
