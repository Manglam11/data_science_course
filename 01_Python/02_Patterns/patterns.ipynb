## Patterns 
### 01. Right-Angled Triangle Pattern

``` 
*
* *
* * *
* * * *
* * * * *
```

````python
# using star
n = 5
for i in range(1, n+1):
    for j in range(i):
        print("*", end=" ")
    print()

print()

# using numbers
for i in range(1, n+1):
    for j in range(i):
        print(j+1, end=" ")
    print()

print()

# continuous numbers
current_number = 1
for i in range(1, n+1):
    for j in range(i):
        print(current_number, end=" ")
        current_number += 1
    print()
````

### 02. Inverted Right-Angled Triangle Pattern

``` 
* * * * *
* * * *
* * *
* *
*
```

````python
# using star
n = 5
for i in range(1, n+1):
    current_inner = n-i+1
    for j in range(1,current_inner+1):
        print("*", end=" ")
       
    print()

print()

# using numbers
for i in range(1, n+1):
    current_inner = n-i+1
    for j in range(1, current_inner+1):
        print(j, end=" ")
       
    print()

print()

# continuous numbers
current_number = 1
for i in range(1, n+1):
    current_inner = n-i+1
    for j in range(1, current_inner+1):
        print(f"{current_number} ", end=" ")
        current_number += 1
    print()
````

### 03. Mirrored Right-Angled Triangle Pattern

``` 
        *
      * *
    * * *
  * * * *
* * * * *
```

````python
# using star
n = 5
for i in range(1, n+1):
    for j in range(1,n-i+1):
        print(" ", end=" ")
    for k in range(1,i+1):
        print("*", end=" ")
       
    print()

print()

# using numbers
for i in range(1, n+1):
    for j in range(1,n-i+1):
        print(" ", end=" ")
    for k in range(1,i+1):
        print(k, end=" ")       
    print()

print()

# continuous numbers
n1=4
current_number = 1
for i in range(1, n1+1):
    for j in range(1,n1-i+1):
        print(" ", end=" ")
    for k in range(1,i+1):
        print(current_number, end=" ")

        current_number += 1
    print()
````

### 04. Pyramid Pattern

``` 
    *
   * *
  * * *
 * * * *
```

````python
# using star
n = 5
for i in range(1, n+1):
    for j in range(1, n-i+1):
        print(" ", end="")
    for k in range(1, i+1):
        print("*",end=" ")
    print()

print()

# using numbers
for i in range(1, n+1):
    for j in range(1, n-i+1):
        print(" ", end="")
    for k in range(1, i+1):
        print(k,end=" ")
    print()

print()

# continuous numbers
n1 = 4
current_number = 1
for i in range(1, n1+1):
    for j in range(1, n1-i+1):
        print(" ", end="")
    for k in range(1, i+1):
        print(current_number,end=" ")
        current_number +=1
    print()

print()
````

### 05. Inverted Pyramid Pattern

``` 
* * * *
 * * *
  * *
   *
```

````python
# using star
n = 5
for i in range(1, n+1):
    for j in range(1, i+1):
        print(" ", end="")
    for k in range(1, n-i+2):
        print("*",end=" ")
    print()

print()

# using numbers
for i in range(1, n+1):
    for j in range(1, i+1):
        print(" ", end="")
    for k in range(1, n-i+2):
        print(k,end=" ")
    print()

print()

# continuous numbers
n1 = 4
current_number = 1
for i in range(1, n1+1):
    for j in range(1, i+1):
        print(" ", end="")
    for k in range(1, n1-i+2):
        print(current_number,end=" ")
        current_number +=1
    print()

print()
````

### 06. Full Pyramid Pattern

``` 
   *
  ***
 *****
*******
```

````python
# using star
n = 5
for i in range(1, n+1):
    for j in range(1, n-i+1):
        print(" ", end="")
    for k in range(0, 2*i-1):
        print("*",end="")
    print()

print()

# using numbers
for i in range(1, n+1):
    for j in range(1, n-i+1):
        print(" ", end="")
    for k in range(0, 2*i-1):
        print(k,end="")
    print()

print()

# continuous numbers
n1 = 4
current_number = 1
for i in range(1, n1+1):
    for j in range(1, n1-i+1):
        print(" ", end="")
    for k in range(0, 2*i-1):
        print(current_number,end="")
        if current_number == 13:
            break
        current_number +=1
    print()

print()
````

### 07. Inverted Full Pyramid Pattern

``` 
*******
 *****
  ***
   *
```

````python
# using star
n = 5
for i in range(1, n+1):
    for j in range(0, i-1):
        print(" ", end="")
    for k in range(0, (n-i+1)*2-1):
        print("*",end="")
    print()

print()

# using numbers
for i in range(1, n+1):
    for j in range(0, i-1):
        print(" ", end="")
    for k in range(0, (n-i+1)*2-1):
        print(k+1,end="")
    print()
````

### 08. Hollow Square Pattern

``` 
*****
*   *
*   *
*   *
*****
```

````python
#using stars
n = 5
for i in range(1, n+1):
    for j in range(1, n+1):
        if j == 1 or j == n or i == 1 or i == n:
            print("*", end="")
        else:
            print(" ", end="")
    print()
````

### 08. Hollow Pyramid Pattern

``` 
    *
   * *
  *   *
 *     *
*********
```

````python
n = 5
for i in range(1, n+1):
    for j in range(1, n-i+1):
        print(" ", end="")
    for k in range(0, 2*i-1):
        if k == 0 or k == 2*i-2 or i == n:
            print("*", end="")
        else:
            print(" ", end="")
    print()
````

### 09. Diamond Pattern

``` 
    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *
```

````python
n = 5
for i in range(1, n+1):
    for j in range(1, n-i+1):
        print(" ", end="")
    for k in range(0, 2*i-1):
        print("*", end="")
    print()
for i in range(1, n):
    for j in range(0, i):
        print(" ", end="")
    for k in range(0, (n-i)*2-1):
        print("*", end="")
    print()
````

### 10. Hollow Diamond Pattern

``` 
    *
   * *
  *   *
 *     *
*********
 *     *
  *   *
   * *
    *
```

````python
n = 5
for i in range(1, n+1):
    for j in range(0, n-i):
        print(" ", end="")
    for k in range(0, 2*i-1):
        if k == 0 or k == 2*i-2 or i == n:
            print("*", end="")
        else:
            print(" ", end="")
    print()
for i in range(1, n):
    for j in range(0, i):
        print(" ", end="")
    for k in range(0, 2*(n-i)-1):
        if k == 0 or k == 2*(n-i)-2:
            print("*", end="")
        else:
            print(" ", end="")
    print()
````

### 11. Butterfly Pattern

``` 
*      *
**    **
***  ***
********
********
***  ***
**    **
*      *
```

````python
n = 5
for i in range (1,n+1):
    for j in range(0, i):
        print("*", end="")
    for k in range(0, 2*(n-i)):
        print(" ", end="")
    for l in range(0, i):
        print("*", end="")
    print()
for i in range (1,n):
    for j in range(0, n-i):
        print("*", end="")
    for k in range(0, 2*i):
        print(" ", end="")
    for l in range(0, n-i):
        print("*", end="")
    print()

````
