### NUMPY Basics
### 1. Indexing & Slicing

```python
import numpy as np

arr = np.array([10,20,30,40,50])
print(arr)
```

```python
# Tasks:
# 1. Get first 3 elements
print(arr[:3:])
# 2. Get last 2 elements
print(arr[:-3:-1])
# 3. Reverse the array
print(arr[::-1])
```

### 5. Array Creation

```python
# Tasks:
# 1. Create array from 1 to 10
arr = np.arange(1,11)
print(arr)
# 2. Create 5 evenly spaced numbers between 0 and 1
arr_2 = np.arange(0, 1, 0.2)
print(arr_2)
# 3. Create 3x3 matrix of zeros
print(np.zeros([3,3], dtype=int))
```

### 6. Statistical Functions

```python
arr = np.array([10, 20, 30, 40])

# Tasks:
# 1. Find mean
print(np.mean(arr))
# 2. Find standard deviation
print(np.std(arr))
# 3. Find min and max
print(np.min(arr))
print(np.max(arr))
```

### 7. Transpose & Ravel

```python
arr = np.array([[1,2,3],[4,5,6]])

# Tasks:
# 1. Transpose matrix
print(np.transpose(arr))
# 2. Convert to 1D array
print(arr.reshape(1,6))
```

### 8. Stacking & Splitting

```python
a = np.array([1,2])
b = np.array([3,4])

# Tasks:
# 1. Stack arrays
print(np.vstack((a,b)))
print(np.hstack((a,b)))
print(np.stack((a,b)))
# 2. Split array into 2 parts
arr = np.arange(16).reshape(4, 4)
upper, lower = np.vsplit(arr, 2)
print(upper)
print(lower)
left, right = np.hsplit(arr,2)
print(left)
print(right)
```

### 9. Dot Product

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Task:
# Find dot product
print(a @ b)
```

### 10. Log, Round Functions

```python
arr = np.array([1, 2, 3])

# Tasks:
# 1. Apply log
print(np.log(arr))
# 2. Round values
data = np.array([1.2345, 2.7891, 5.5555])
print(np.round(data, 2))
print(np.floor(data))
print(np.ceil(data))
print(np.trunc(data))
```
