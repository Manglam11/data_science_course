### 1. List Creation

```python
l1 = [1, 2, 50, 60, 100]
print(l1)
```

### 2. Indexing

```python
print(l1[0]) #print first element
print(l1[2]) #print third element
print(l1[-1]) #print last element
```

### 3. Slicing

```python
print(l1[0:3]) #print first three element
print(l1[-2:]) #print last 2 element
print(l1[::-1]) #print list in reverse
```

### 4. Add Items

```python
l2 = ["apple", "banana"]
l2.append("mango") #Add "mango" at the end
l2.insert(1, "orange") #Insert "orange" at index 1
print(l2)
```

```python
l3 = [1, 2, 3, 4, 5]
l3.remove(3) #Remove 3
l3.pop(-1)#Remove last element
print(l3)
```

### 6. Change Items

```python
l4 = ["red", "blue", "green"]
l4[1] = "yellow" #Change "blue" to "yellow"
print(l4)
```

### 7. Copy List

```python
l2_copy = l2.copy()
l2_copy.append("papaya") #Create a copy of a list and modify the copy.
l2_copy.insert(-1, l4)
print(l2_copy)
print(l2) #Check: original list should NOT change.
```

### 8. Join Lists

```python
a = [1, 2]
b = [3, 4]

c = a+b #Join into one list.
print(c)
```

### 10. Reverse & Sort

```python
nums = [5, 2, 8, 1]
nums.sort() #Sort ascending
print(nums)
nums.sort(reverse=True) #Sort descending
print(nums)
nums.reverse() #Reverse the list
print(nums)
```

### 11. Loop Through List

```python
for i in l2: #Print all elements
    print(i)
```

### 12. Count

```python
nums = [1, 2, 2, 3, 2, 4] #Count how many times 2 appears.
print(nums.count(2))
```

### 13. all() and any()

```python
nums = [2, 4, 6, 8]
all_result = all(num % 2 ==0 for num in nums) #Check if all numbers are even
print(all_result)
any_result = any(num>5 for num in nums)
print(any_result) #Check if any number is greater than 5
```
