### 1. Basic Practice

### 2. Tuple Indexing

```python
t2 = (10, 20, 30, 40, 50)
print(t2[0]) #Print the first element
print(t2[-1]) #Print the last element
print(t2[2]) #Print the third element
```

### 3. Tuple Slicing

```python
print(t2[1:4]) #Print elements from index 1 to 3
print(t2[-1:1:-1]) #Print last 3 elements
print(t2[::-1]) #Reverse the tuple using slicing
```

### 4. Loop Through Tuple

```python
t3 = ("apple", "banana", "cherry")
for item in t3:
    print(item)
```

### 5. Tuple Membership

```python
print("banana" in t3)
```

### 6. Find Index Position

```python
t4 = (5, 10, 15, 20, 25)
print(t4.index(15))
```

### 7. Sorting Tuple

```python
t5 = (4, 1, 3, 2)
print(sorted(t5))
```

### 8. Count Even Numbers

```python
t6 = (1, 2, 3, 4, 5, 6)
even_numbers = sum(num%2==0 for num in t6)
print(even_numbers)
```

### 9. Tuple to List

```python
t7 = (10, 20, 30)
list_ob = list(t7)
print(type(list_ob))
print(list_ob)
```

### 10. Remove Duplicates

```python
t8 = (1, 2, 2, 3, 4, 4, 5)
clean_tuple = tuple(set(t8))
print(clean_tuple)
```

### 11. Frequency Count

```python
t9 = (1, 2, 3, 3, 3, 4)
for item in set(t9):
    print(f"{item} -> {t9.count(item)} times")
```
